```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import re

X = X.drop(columns=['Account', 'Account.1'])

# Convert string to datetime
X['timestamp_column_name'] = pd.to_datetime(X['Timestamp'])

# Extract useful numeric features
X['hour'] = X['timestamp_column_name'].dt.hour
X['day'] = X['timestamp_column_name'].dt.day
X['weekday'] = X['timestamp_column_name'].dt.weekday
X['month'] = X['timestamp_column_name'].dt.month

X = X.drop(['Timestamp'], axis=1)

categorical_cols = ['Receiving Currency', 'Payment Currency', 'Payment Format']

encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_array = encoder.fit_transform(X[categorical_cols])

encoded_df = pd.DataFrame(
    encoded_array,
    columns=encoder.get_feature_names_out(categorical_cols),
    index=X.index  # preserve index for joining later
)

X = pd.concat([X.drop(columns=categorical_cols), encoded_df], axis=1)

import re

def encode_account_grouped_flexible(val):
    val = str(val).strip()
    
    # Extract all letters and digits separately
    letters = ''.join(re.findall(r'[A-Za-z]', val))
    digits_str = ''.join(re.findall(r'\d+', val))  # Collect only digits

    # Default: handle at most 2 letters
    l1 = ord(letters[0].upper()) - ord('A') if len(letters) >= 1 else -1
    l2 = ord(letters[1].upper()) - ord('A') if len(letters) >= 2 else -1

    try:
        digits = int(digits_str)
    except ValueError:
        digits = -1  # fallback if digits can't be parsed

    return pd.Series({'AG_L1': l1, 'AG_L2': l2, 'AG_Num': digits})


X[['AG_L1', 'AG_L2', 'AG_Num']] = X['Account_Grouped'].apply(encode_account_grouped_flexible)
X[['AG1_L1', 'AG1_L2', 'AG1_Num']] = X['Account1_Grouped'].apply(encode_account_grouped_flexible)
X.drop(columns=['Account_Grouped', 'Account1_Grouped'], inplace=True)

X
```

# Now, Oversample (SMOTE) and Undersample Majority Class.

- We will get new balanced data -- X and y.

```python
# Identify datetime columns
datetime_cols = X.select_dtypes(include=['datetime64']).columns
print(datetime_cols)

# Convert them into numeric features (like year, month, etc.)
for col in datetime_cols:
    X[f'{col}_year'] = X[col].dt.year
    X[f'{col}_month'] = X[col].dt.month
    X[f'{col}_day'] = X[col].dt.day
    X[f'{col}_minute'] = X[col].dt.minute
    # drop original datetime column
    X.drop(columns=col, inplace=True)
    
X
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
from imblearn.under_sampling import EditedNearestNeighbours
from sklearn.utils import shuffle

# https://imbalanced-learn.org/stable/references/generated/imblearn.combine.SMOTEENN.html
def get_smoteenn():
    return SMOTEENN(
        sampling_strategy='auto',
        random_state=42,
        smote=SMOTE(k_neighbors=3, n_jobs=-1),
        enn=EditedNearestNeighbours(n_neighbors=3, n_jobs=-1),
        n_jobs=-1
    )

def batchwise_smoteenn(X, y, batch_size=200000):
    Xy_resampled_batches = []

    # Shuffle first to randomize the order
    X, y = shuffle(X, y, random_state=42)

    num_rows = len(X)
    num_batches = (num_rows + batch_size - 1) // batch_size

    for i in range(num_batches):
        print(f"Processing batch {i+1}/{num_batches}...")

        # Define batch boundaries
        start = i * batch_size
        end = min((i + 1) * batch_size, num_rows)

        X_batch = X.iloc[start:end]
        y_batch = y.iloc[start:end]

        # Apply SMOTEENN to this batch
        sme = get_smoteenn()
        try:
            X_resampled_batch, y_resampled_batch = sme.fit_resample(X_batch, y_batch)
            batch_df = pd.concat([pd.DataFrame(X_resampled_batch), pd.Series(y_resampled_batch, name='target')], axis=1)
            Xy_resampled_batches.append(batch_df)
        except Exception as e:
            print(f"Skipping batch {i+1} due to error: {e}")

    # Concatenate all batch results
    final_df = pd.concat(Xy_resampled_batches, ignore_index=True)
    return final_df.drop(columns='target'), final_df['target']


# fit_resample (X, y, **params) : Resample the dataset.
X, y = batchwise_smoteenn(X, # X : {array-like, dataframe, sparse matrix} of shape (n_samples, n_features)
                              y, # y : array-like of shape (n_samples,)
                              batch_size=200000) 

X
```

# Pearson Co-relation.

```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import sparse
from sklearn.preprocessing import OneHotEncoder

y_array = y.to_numpy()

# ----- Step 2: Compute Pearson Correlation (Sparse-Safe) -----
correlations = []
for i in tqdm(range(X_final_sparse.shape[1]), desc="Calculating Pearson correlation"):
    col = X_final_sparse.getcol(i).toarray().ravel()
    if np.std(col) == 0:
        corr = 0
    else:
        corr = np.corrcoef(col, y_array)[0, 1]
    correlations.append(corr)

# ----- Step 3: Prepare Results -----
encoded_feature_names = encoder.get_feature_names_out(categorical_cols)
all_feature_names = list(encoded_feature_names) + list(numeric_X.columns)

corr_df = pd.DataFrame({
    'feature': all_feature_names,
    'correlation': correlations
}).dropna()
```

```python
top_100 = corr_df.reindex(corr_df['correlation'].abs().sort_values(ascending=False).index).head(100)

# ----- Step 4: Plot -----
plt.figure(figsize=(12, 18))
sns.barplot(data=top_100, y='feature', x='correlation', palette='coolwarm')
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Top 100 Pearson Correlated Features with Target (Is Laundering)')
plt.tight_layout()
plt.show()
```

```python
bottom_100 = corr_df.reindex(corr_df['correlation'].sort_values(ascending=False).index).tail(100)

# ----- Step 4: Plot -----
plt.figure(figsize=(12, 18))
sns.barplot(data=bottom_100, y='feature', x='correlation', palette='coolwarm')
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Bottom to Top 100 Pearson Correlated Features with Target (Is Laundering)')
plt.tight_layout()
plt.show()
```
