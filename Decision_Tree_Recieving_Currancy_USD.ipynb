# Build decision tree models


```python
# (Code efrom HW-2, 3, and 4)

# from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.metrics import roc_auc_score, make_scorer
import numpy as np
import gc

# k_fold_res list.
neigh_scores_k_fold_res = {}

for name, clf in clfs.items():
    neigh_scores = cross_val_score(clf, # estimator : estimator object implementing ‘fit’.
                                                   # predict_cols = dataFile.drop(columns=['classification'])
                                                   # X
                                                   X_normalized, 
                                                   # target_col = dataFile['classification']
                                                   # y
                                                   y, 
                                                   # n_splits = 10 in the above function.
                                                   # cv : int, cross-validation generator or an iterable, default=None
                                                   cv=cross_validation, 
                                                   # we just need roc_auc_score, otherwise add all.
                                                   # A str (see The scoring parameter: defining model evaluation rules) or a scorer callable object / function with signature scorer(estimator, X, y) which should return only a single value.
                                                   scoring=make_scorer(roc_auc_score, needs_proba=True, multi_class="ovr"),
                                                   # n_jobsint, default=None. Number of jobs to run in parallel. Training the estimator and computing the score are parallelized over the cross-validation splits. None means 1 unless in a joblib.parallel_backend context. -1 means using all processors. See Glossary for more details.
                                                   # mean to get (mean) of all 10 folds as asked.
                                                   n_jobs=4,
                                                   verbose=0)

    # Store results
    neigh_scores_k_fold_res[clf] = np.mean(neigh_scores)
    del neigh_scores
    gc.collect()
      
# SHow the list and their AUC values.        
neigh_scores_k_fold_res
# Convert results to DataFrame and display
# temp_results_df = pd.DataFrame(k_fold_res)
# temp_results_df (Next roew)
```

```python
# Create a new dataframe temp_results_df.
temp_results_df = pd.DataFrame(list(neigh_scores_k_fold_res.items()), # List the set items.
                          # Column name ["DecisionTreeClassifier_Model", "AUC Score"]
                          columns=["DecisionTreeClassifier_Model", "AUC Score"],
                          dtype=None,
                          copy=None)

# Display the temp_results_df.
display("temp_results_df score is: \n")
display(temp_results_df)
```

```python
# Find the best model based on the highest AUC score, find the location here.
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html
# df.loc['viper']
# max_speed    4
# shield       5
# Name: viper, dtype: int64
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html
temp_best_model_name = temp_results_df.loc[temp_results_df["AUC Score"].idxmax(axis=0, # axis : {0 or ‘index’, 1 or ‘columns’}, default 0
                                                                               skipna=True), # skip : nabool, default True
#                                                                                numeric_only=False), # numeric_only : bool, default False
                                           "DecisionTreeClassifier_Model"]

# Fine the highest AUC score.
score_temp_model_selected = temp_results_df["AUC Score"].max(skipna=True, # skipna : bool, default True
                                               numeric_only=False) # numeric_only : bool, default False

# Display the best model and its score from k_fold_res.
# Model-5 is the best. Model-5: DecisionTreeClassifier_MinSamplesSplit10_criterionGini
print("********************************************************************************************")
print("1. THE temp_best_model_name is := ", temp_best_model_name)
print("********************************************************************************************")
print("2. THE score_temp_model_selected is := ", score_temp_model_selected)
print("********************************************************************************************")
```
