# Things I did for X.

- Convert Timestamp into columns like i.e., {hour, day, weekday, and month}.
- For Account_Grouped and Account1_Grouped columns,
    - Extract first two letters (if present)
    - Collect the digits at the end (whatever part is numeric) 

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import re

# Drop redundant columns
X = X.drop(columns=['Account', 'Account.1'])

# Convert string to datetime
X['timestamp_column_name'] = pd.to_datetime(X['Timestamp'])

# Extract datetime features
X['hour'] = X['timestamp_column_name'].dt.hour
X['day'] = X['timestamp_column_name'].dt.day
X['weekday'] = X['timestamp_column_name'].dt.weekday
X['month'] = X['timestamp_column_name'].dt.month

X = X.drop(['Timestamp'], axis=1)

# One-hot encode selected categorical columns
categorical_cols = ['Receiving Currency', 'Payment Currency', 'Payment Format']

encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_array = encoder.fit_transform(X[categorical_cols])

encoded_df = pd.DataFrame(
    encoded_array,
    columns=encoder.get_feature_names_out(categorical_cols),
    index=X.index
)

X = pd.concat([X.drop(columns=categorical_cols), encoded_df], axis=1)

def fast_encode_account_grouped(df, column_name, prefix):
    col = df[column_name].astype(str).str.strip()

    # Extract first two letters
    letters = col.str.extract(r'([A-Za-z])?([A-Za-z])?')
    digits = col.str.extract(r'(\d+)')

    l1 = letters[0].str.upper().apply(lambda x: ord(x) - ord('A') if pd.notnull(x) else -1)
    l2 = letters[1].str.upper().apply(lambda x: ord(x) - ord('A') if pd.notnull(x) else -1)
    num = digits[0].fillna(-1).astype(int)

    return pd.DataFrame({
        f'{prefix}_L1': l1,
        f'{prefix}_L2': l2,
        f'{prefix}_Num': num
    }, index=df.index)

# Apply vectorized encoding
X = pd.concat([
    X,
    fast_encode_account_grouped(X, 'Account_Grouped', 'AG'),
    fast_encode_account_grouped(X, 'Account1_Grouped', 'AG1')
], axis=1)

X.drop(columns=['Account_Grouped', 'Account1_Grouped'], inplace=True)
X
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

# Isolation Forest


```python
# Step 2: Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
labels_if = iso_forest.fit_predict(X)
```

# Autoencoder (using Keras)

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

input_dim = X.shape[1]
input_layer = Input(shape=(input_dim,))
encoded = Dense(16, activation="relu")(input_layer)
encoded = Dense(8, activation="relu")(encoded)
decoded = Dense(16, activation='relu')(encoded)
output_layer = Dense(input_dim, activation='linear')(decoded)

autoencoder = Model(inputs=input_layer, outputs=output_layer)
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.fit(X, X, epochs=20, batch_size=32, shuffle=True, verbose=0)

reconstructions = autoencoder.predict(X)
reconstruction_error = np.mean((X - reconstructions)**2, axis=1)
threshold = np.percentile(reconstruction_error, 95)
labels_auto = np.where(reconstruction_error > threshold, -1, 1)
```

# Step 5: KMeans Clustering

```python
kmeans = KMeans(n_clusters=2, random_state=42)
cluster_labels = kmeans.fit_predict(X)
distances = np.linalg.norm(X - kmeans.cluster_centers_[cluster_labels], axis=1)
threshold_kmeans = np.percentile(distances, 95)
labels_kmeans = np.where(distances > threshold_kmeans, -1, 1)
```

# PCA, TSNE, UMAP

```python
from sklearn.manifold import TSNE
import umap.umap_ as umap

pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# t-SNE
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_tsne = tsne.fit_transform(X)

# UMAP
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X)
```

# Output


```python
import matplotlib.pyplot as plt

def plot_anomalies(X_2d, labels, title, normal_label=0, anomaly_label=1, 
                   x_label="PCA Component 1", y_label="PCA Component 2", save_path=None):
    plt.figure(figsize=(6, 4))

    # Plot normal points with low opacity
    plt.scatter(X_2d[labels == normal_label, 0], X_2d[labels == normal_label, 1], 
                c='blue', label='Normal', s=20, alpha=0.5)

    # Plot anomaly points with higher opacity
    plt.scatter(X_2d[labels == anomaly_label, 0], X_2d[labels == anomaly_label, 1], 
                c='red', label='Anomaly', s=8, alpha=0.05, edgecolors='black', linewidths=0.2)

    # Count labels
    normal_count = (labels == normal_label).sum()
    anomaly_count = (labels == anomaly_label).sum()

    # Title and axis labels
    plt.title(f"{title} (Normal: {normal_count}, Anomaly: {anomaly_count})")
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend(loc='best', frameon=True)
    plt.grid(True)
    plt.tight_layout()

    # Save if path is provided
    if save_path:
        plt.savefig(save_path, dpi=300)

    plt.show()

# ---- PLOT RESULTS ----
# PCA
plot_anomalies(X_2d, labels_if, "Isolation Forest (PCA)", normal_label=1, anomaly_label=-1, x_label="PCA 1", y_label="PCA 2")
plot_anomalies(X_2d, labels_auto, "Autoencoder (PCA)", normal_label=0, anomaly_label=1, x_label="PCA 1", y_label="PCA 2")
plot_anomalies(X_2d, labels_kmeans, "KMeans Clustering (PCA)", normal_label=1, anomaly_label=-1, x_label="PCA 1", y_label="PCA 2")

# t-SNE
plot_anomalies(X_tsne, labels_if, "Isolation Forest (t-SNE)", normal_label=1, anomaly_label=-1, x_label="t-SNE 1", y_label="t-SNE 2")
plot_anomalies(X_tsne, labels_auto, "Autoencoder (t-SNE)", normal_label=0, anomaly_label=1, x_label="t-SNE 1", y_label="t-SNE 2")
plot_anomalies(X_tsne, labels_kmeans, "KMeans Clustering (t-SNE)", normal_label=1, anomaly_label=-1, x_label="t-SNE 1", y_label="t-SNE 2")

# UMAP
plot_anomalies(X_umap, labels_if, "Isolation Forest (UMAP)", normal_label=1, anomaly_label=-1, x_label="UMAP 1", y_label="UMAP 2")
plot_anomalies(X_umap, labels_auto, "Autoencoder (UMAP)", normal_label=0, anomaly_label=1, x_label="UMAP 1", y_label="UMAP 2")
plot_anomalies(X_umap, labels_kmeans, "KMeans Clustering (UMAP)", normal_label=1, anomaly_label=-1, x_label="UMAP 1", y_label="UMAP 2")
```
